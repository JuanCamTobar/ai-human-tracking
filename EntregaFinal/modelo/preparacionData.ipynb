```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_time_series_with_outliers(df, outliers_mask, columns, title='Time series with outliers'):
    plt.figure(figsize=(14,6))
    for col in columns:
        plt.plot(df.index, df[col], label=col)
        # Marca los outliers en rojo
        plt.scatter(df.index[outliers_mask], df.loc[outliers_mask, col], color='red', s=30)
    plt.legend()
    plt.title(title)
    plt.xlabel('Index / Time')
    plt.ylabel('Value')
    plt.show()
```

```python
df = df_agacharse
mask_outliers = (np.abs(stats.zscore(df.select_dtypes(include=[float]), nan_policy='omit')) > 3).any(axis=1)

cols_to_plot = ['landmark_11_x','landmark_11_z','landmark_12_z','landmark_12_x', 'landmark_23_y', 'landmark_24_y','cara_prom_x','cara_prom_y','cara_prom_z','mano_izquierda_prom_x','mano_derecha_prom_x']

plot_time_series_with_outliers(df, mask_outliers, cols_to_plot)

```
